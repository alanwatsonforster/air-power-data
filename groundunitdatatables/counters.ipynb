```python
import glass
import glass.draw

import math
import os

def drawcounter(
    type, color="white", sighted=False, identified=False, azimuth=None, protectionclass=None
):

    glass.startgamesetup(None, sheets=[["A"]], writelogfiles=False)

    E = glass.setupgroundunit(
        "",
        type=type,
        hexcode="origin",
        color=color,
        counter=True,
        sighted=sighted,
        identified=identified,
        azimuth=azimuth
    )

    groundunitcountersize = 0.6 * math.sqrt(4 / 3)
    dotspercounter = 50
    dotsperhex = dotspercounter / groundunitcountersize

    glass.draw.startcanvas(
        -0.5 * groundunitcountersize,
        -0.5 * groundunitcountersize,
        +0.5 * groundunitcountersize,
        +0.5 * groundunitcountersize,
        dotsperhex=dotsperhex,
    )

    E._draw()

    glass.draw.writecanvastofile("counters/%s/%s.png" % (color, type))


drawcounter("infantry platoon", color="tan", sighted=True)
```
