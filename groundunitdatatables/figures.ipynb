```python
def prettyname(name, counter):
    if counter:
        name = name.replace(" ", "\n  ")
        #name = name.replace("\n  squad", "")
        #name = name.replace("\n  section", " ")
        #name = name.replace("\n  platoon", "  ")
        #name = name.replace("\n  battery", "   ")
    else:
        name = name.replace(" squad", "")
        name = name.replace(" section", " ")
        name = name.replace(" platoon", "  ")
        name = name.replace(" battery", "   ")        
    return name
```
